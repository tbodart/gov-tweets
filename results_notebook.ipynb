### *How actively are United States governers sharing information about COVID via Twitter?*  
Examining various aspects of official US governor Twitter accounts to gain insights into how and where information is being shared.

### Data Collection  
I used the REST Api from tweepy to run a user search for each US governor and collected up to 100 tweets per governor from April 23rd to April 30th, filtering out retweets with the search term `-filter:retweets`. I pulled their verified Twitter accounts from Twitter's official list of governor accounts (https://twitter.com/i/lists/88692902/members) and created a dictionary `governor_dict.py`. I then saved various aspects of their tweets &#151; username, followers, full text, and hashtags, to name a few &#151; to a csv named "tweets.csv". This process is saved in `gather_tweets.py`.

### Data Processing  
To process the data I created functions to generate specific data for each state's governor and compiled a new DataFrame containing the processed data, `processed_df` and wrote it to `part3_processed_data.csv`. The characteristics of interest for each governor were: total and average number of times they mentioned the words "covid", "vaccine", or "pandemic" in their tweets, the proportion of their tweets which mentioned the pandemic, the most common hashtag they used, and when their Twitter was created. This process is saved in `data_process.py`.
